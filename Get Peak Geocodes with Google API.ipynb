```python
peaks=pd.read_csv("peaks.csv")
peaks_df=pd.DataFrame(peaks)
peaks.head(5)
```

```python
peaks["lat"]=""
peaks["long"]=""
```

For each peak, use Google Geocode API to fill in latitude and longitude

```python
for index in range(len(peaks["pkname"])):    
    peak=peaks.at[index,"pkname"]
    print("index:"+str(index)+", Peak:"+peak)
    try:
        query_url="https://maps.googleapis.com/maps/api/geocode/json?address="+peak+"&key="+api_key
        response=req.get(query_url).json()    
        peaks.at[index,"lat"]=response["results"][0]["geometry"]["location"]["lat"]
        peaks.at[index,"long"]=response["results"][0]["geometry"]["location"]["lng"]
        print("Found")
    except IndexError:
        peaks.at[index,"lat"]="NotFound"
        peaks.at[index,"long"]="NotFound"
        print("Not Found")
```
